```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
urls = ['https://www.amazon.eg/s?bbn=21832868031&rh=n%3A21832868031%2Cp_89%3ASAMSUNG&dc&language=en&qid=1657961820&rnid=22541269031&ref=lp_21832868031_nr_p_89_2',
       'https://www.amazon.eg/-/en/s?i=electronics&bbn=21832868031&rh=n%3A21832868031%2Cp_89%3ASAMSUNG&dc&page=3&language=en&qid=1658020856&rnid=22541269031&ref=sr_pg_3']
```

```python
f = open("amazon-samsung2.csv", "w")
header = "properties, price, rate, image\n"
f.write(header)

for url in urls:
    client = urlopen(url)
    html = client.read()
    client.close()
    soup = bs(html, "html.parser")
    containers = soup.find_all("div",{"class": "sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20"})
   
    for container in containers:
    
        name = container.findAll("a",{"class": "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
        properties = name[0].text.replace(',','|').strip().lstrip()

        try:       
            pricee = container.findAll("span",{"class": "a-price-whole"})
            price = pricee[0].text.replace(',','.')+'EGY'.strip()
        except IndexError :
            price = 'NaN'
    
        try:      
            ratee = container.findAll("a",{"class":"a-popover-trigger a-declarative"})
            rate = ratee[0].text
        except IndexError:
            rate = 'NaN'

        img = container.findAll("img",{"class":"s-image"})
        image = img[0]['src']
    
        f.write(properties + ", " + price + ", " + rate + ", " + image + "\n")
f.close()
```
